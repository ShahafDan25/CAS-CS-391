Next, we use divide our expression to composite functions and take the appropriate derivatives by assigning:

$g(x) = x^4$ and $f(g) = 3 \cdot log(g)$

therefore:
$$ ... = \frac{df}{dg} \cdot \frac{dg}{dx} + -2x^{-3} = \frac{3}{g \cdot ln(10)} \cdot 4x^3 - \frac{2}{x^3} = \boxed{\frac{12}{x \cdot ln(10)} - \frac{2}{x^3}}$$

**1.2** $\frac{df}{dx} x^2y + e^{x^2+y} + x\log(y^2+y) = ?$

**Answer:**

We are asked to take the derivative with the respect to x as follows:
$$\frac{d}{dx} [x^2y + e^{x^2+y} + x\log(y^2+y)] = ...$$

First: note that as seen in piazza, we are allowed to consider the log in the function as the natural log (ln) which has base e.

Since we are derivaiting the expression with respect to X, and any other variables (y in this case) is not considered as a function of x, we can consider these other variables as coefficients. therefore:

1) $\frac{d}{dx} [x^2y] = 2yx$

2) $\frac{d}{dx}[e^{x^2 + y}] = 2x\cdot e^{x^2+y}$

3) $\frac{d}{dx} [x \cdot ln(y^2 + y)] = ln(y^2 + y)$

Therefore, we can conclude that

$$ ... = \boxed{2yx + 2x\cdot e^{x^2+y} + ln(y^2 + y)}$$

**1.3** $\frac{df}{dy} x^2y + e^{x^2+y} + x\log(y^2+y) = ?$

**Answer:**
We are asked to take the derivative with the respect to y as follows:
$$\frac{d}{dy} [x^2y + e^{x^2+y} + x\log(y^2+y)] = ...$$

In order to do that, let's divide the expression to 3 distinguished segments and take the derivative with respect to x of each one of them as follows:

1) $\frac{d}{dy} [x^2y] = x^2$

2) $\frac{d}{dy}[e^{x^2 + y}] = e^{x^2+y}$

3) $\frac{d}{dy} [x \cdot ln(y^2 + y)] = x\cdot \frac{2y + 1}{y^2 + y}$

Therefore, we can conclude that

$$ ... = \boxed{x^2 + e^{x^2+y} + x\cdot \frac{2y + 1}{y^2 + y}} $$

Let's look at the Ames Housing data set, which contains data about homes sold in Ames, Iowa between 2006-2010.  It's a large data set, and we will only be using a small subset of features.  You can read more about the full data set here:  http://jse.amstat.org/v19n3/decock.pdf

**2.2** Let's run the experiment!  Predict sales price using square feet, year built, and year remodeled.  First a linear regression on the data, and then a quadratic regression.  Calculate the mean squared error for each model (they will be huge!)  Which one performs better?

```python
# Answer linear regression:

from sklearn.linear_model import LinearRegression

model = LinearRegression()
X = ames[['sqft', 'year_built', 'year_remod']]
y = ames.sale_price
model.fit(X, y)
prediction = model.predict(X)

from sklearn.metrics import mean_squared_error
print("Linear Regression's means squared error:")
mean_squared_error(prediction, y)
```

```python
#Answer quadratic (polynomial) regression:

from sklearn.preprocessing import PolynomialFeatures
import numpy as np
X = ames[['sqft', 'year_built', 'year_remod']]
y = ames.sale_price

model = PolynomialFeatures(2) #degree = 2 means it is quadratic regression
X_poly = model.fit_transform(X) # set the x matrix for the 2nd degree polynomial regression
model_lin = LinearRegression()
model_lin.fit(X_poly, y) #create the fit of the matrix
prediction = model_lin.predict(X_poly) # set the prediction
print("Polynomial Regression's means squared error:")
mean_squared_error(prediction, y)
```

**Answer:**
We see that the polynomial regression's mean squared error is lower than the linear regression's (though they are both extremely large). Therefore we can conclude that the polynomial regression performed better (due to the fact it's mean squared error is smaller).
