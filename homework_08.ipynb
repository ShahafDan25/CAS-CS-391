Let's review the following calculation:

First, let's begin by assigning the following events:
A = Diagnosed Pneumonia, B = Diagnosed Flu, C = Diagnosed Healthy, D = Having both Fever and Headache

Accoprdingly, we will calculate their probabilities from the data provided to us:
$$ Pr(A) = \frac{10}{100} = \frac{1}{10} \\ Pr(B) = \frac{20}{100} = \frac{2}{10} \\ Pr(C) =  \frac{70}{100} = \frac{7}{10} \\ Pr(D) = \frac{5 + 9 + 2}{100} = \frac{16}{100}$$

Next, We can calculate the conditional probabilities as follows:
$$Pr(D | A) = \frac{Pr(D\,\cap\,A)}{Pr(A)} \\ Pr(D | B) = \frac{Pr(D\,\cap\,B)}{Pr(B)} \\ Pr(D | C) = \frac{Pr(D\,\cap\,C)}{Pr(C)}$$

Therefore, we can next plug those into each scenario's individual bayes theorem forumla. We shall begin be recalling the Bayes theorem formula:

$ \quad Pr(E | X) = \frac{Pr(X | E) \cdot Pr(E)}{Pr(X)}$

Therefore:

$$\quad Pr(A | D) = \frac{Pr(D | A) \cdot Pr(A)}{Pr(D)} = \frac{\frac{5}{10}\cdot \frac{1}{10}}{\frac{16}{100}} = \frac{5 / 100}{16 / 100} = \frac{5}{16}\\\quad Pr(B | D) = \frac{Pr(D | B) \cdot Pr(B)}{Pr(D)} = \frac{\frac{9}{20} \cdot \frac{2}{10}}{\frac{16}{100}} =\frac{P\frac{18}{200}}{\frac{16}{100}} = \frac{9}{16} \\\quad Pr(C | D) = \frac{Pr(D | C) \cdot Pr(C)}{Pr(D)} = \frac{\frac{2}{70} \cdot \frac{7}{10}}{\frac{16}{100}} = \frac{\frac{14}{700}}{\frac{16}{100}} = \frac{\frac{2}{100}}{\frac{16}{100}} = \frac{1}{8}$$
