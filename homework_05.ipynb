**1.1** Let $A = \begin{bmatrix} a_{11} & 0 & \cdots & 0\\0 & a_{22}& & 0\\ \vdots & & \ddots & \\ 0 & 0&\cdots & a_{nn} \end{bmatrix}$ be a diagonal matrix. Find the SVD decomposition of $A$. Based on the decomposition, formulate a general rule for the SVD of a diagonal matrix.

**Answer:**
$$ A_{n x n} = U_{n x n} \Sigma_{n x n} V^T_{n x n}$$
Lets begin by acknowledging the need to compute $AA^T$ to calculate V and $A^TA$ to calculate U.
Notice, however, that $A^T = A$, since A is a diagonal matrix.

Therefore:
$$ AA^T = A^TA = AA = \begin{bmatrix} a^2_{11} & 0 & \cdots & 0\\0 & a^2_{22}& & 0\\ \vdots & & \ddots & \\ 0 & 0&\cdots & a^2_{nn} \end{bmatrix} $$

To find the eigenvectors, we must perform:
$$ AA^T - \lambda I = \begin{bmatrix} a^2_{11} & 0 & \cdots & 0\\0 & a^2_{22}& & 0\\ \vdots & & \ddots & \\ 0 & 0&\cdots & a^2_{nn} \end{bmatrix} - \lambda \cdot \begin{bmatrix} 1 & 0 & \cdots & 0\\0 & 1& & 0\\ \vdots & & \ddots & \\ 0 & 0&\cdots & 1 \end{bmatrix} = \begin{bmatrix} a^2_{11} - \lambda& 0 & \cdots & 0\\0 & a^2_{22}- \lambda& & 0\\ \vdots & & \ddots & \\ 0 & 0&\cdots & a^2_{nn}- \lambda \end{bmatrix}$$

Therefore, we can conlude that the eigenvalues are $\lambda = a_{11}^2, a_{11}^2,..., a_{nn}^2 $

Hence, we get that 

$$ U = V = \begin{bmatrix} 1 & 0 & \cdots & 0\\0 & 1& & 0\\ \vdots & & \ddots & \\ 0 & 0&\cdots & 1 \end{bmatrix}$$,
which can therefore help us understand that 

$$ \Sigma = A = \begin{bmatrix} a^2_{11} & 0 & \cdots & 0\\0 & a^2_{22}& & 0\\ \vdots & & \ddots & \\ 0 & 0&\cdots & a^2_{nn} \end{bmatrix}$$

In conclusion, we can sum up two main rules for a diagnoal matrix such as A and its SVD:

1) $\Sigma = A$

2) $ U = V = $ diagonal identity matrix of size n

**1.3** Let $A = U \Sigma V^T$ be the SVD decomposition of $A$. Denote by $u_1$ the first column of $U$ and by $\sigma_1$ the corresponding eigen value. Show that $|u_1^T A | = \sigma_1$.

**Answer:**

Consider the following: 

$A = U \Sigma V^T$ is the SVD decomposition of A

U and V are both orthogonal matrices, $\Sigma$ is a diagnoal matrix.

U is an m by n matrix, such that U can be computed by $\sigma u_i = A v_i$ or $u_i = \frac{1}{\sigma} A v_i$.

Therefore, we can rearrange the A matrix such that 
$$A = (u_1 \sigma_1 v_1)^T + (u_2 \sigma_2 v_2)^T + ... + (u_i \sigma_i v_i)^T$$ 

In which, we can identify that $\sigma_1$ is a singular value and $\sigma_1^2 is an eigenvalue$

Therefore, since $\frac{A v_1}{u_1} = \sigma_1$ and $\frac{v_1}{u_1}= u_1^T$

Then we can conclude that V has a size of n by n, and therefore

$$ \boxed{|u_1^T A | = \sigma_1} $$

### **2.3**

Now, let's compute a rank-5 approximation of our dataset.  To do this, compute the SVD your data matrix.  This should give you 9 singular values in your matrix $\Sigma$.  Keep the first five singular values as they are, but change the last five singular values to 0.  Then, multiply $U \Sigma V^T$ back together again.  This will create a rank-5 projection that you can use in **2.4**. Write your code here to create the rank-5 SVD. (will not be graded.)

```python
# Answer:
U,S,VT = svd(normalized) #compute SVD
for x in [5,6,7,8]:
    S[x] = 0

newS = np.zeros((normalized.shape[0], normalized.shape[1]))
S = np.diag(S)

newS = np.concatenate((S, np.zeros([normalized.shape[0] - S.shape[0], S.shape[1] ])), axis =0)
print(newS.shape)

rank_five = U.dot(newS.dot(VT))

print(rank_five)
```
